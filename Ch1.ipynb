## groupby simple

```python
group_year_df = df.groupby("year")
mean = group_year_df["lifeExp"].mean()
mean
```

## groupby multiple

```python
multiple_group = df.groupby(["year", "continent"])
mean = multiple_group[["lifeExp", "gdpPercap"]].mean()
mean
```

```python
df["country"].unique().size
```

```python
group_year = df.groupby("year")
group_year["lifeExp"].mean().plot(
    xlabel="Year", 
    ylabel="Mean life exp", 
    title="Mean life exp per year"
)
```
